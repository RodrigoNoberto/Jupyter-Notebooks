# Exploratory Data Analysis using Pandas, Matplotlib and SQL(pandasql)

```python
# importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandasql import sqldf
```

```python
# importing the dataframe
df = pd.read_csv('automobileEDA.csv')
df.head()
```

```python
# querying the audi's cars
q = """select * from df where make = 'audi' """
sqldf(q)
```

```python
# Car makes where the price of the veicules is bigger than 20,000
q = """select distinct(make) from df where price > 20000 """
sqldf(q)
```

```python
# shape of the dataframe / 201 rows and 29 columns
df.shape
```

```python
# top 10 most expensive cars
q = """select * from df order by price desc limit 10"""
sqldf(q)
```

```python
# average price of every make
q = """select make, avg(price) as 'avg_price' from df group by make order by avg(price) desc"""
sqldf(q)
```

```python
# types of every column
df.dtypes
```

```python
plt.scatter(df['price'],df['engine-size'])
plt.xlabel('Price')
plt.ylabel('Engine-size')
plt.show()
```

```python
plt.scatter(df['price'],df['horsepower'],color='g')
plt.xlabel('Price')
plt.ylabel('Horsepower')
plt.show()
```

```python
# new dataframe with only 'make' and 'price' columns from the original one
df2 = df[['make', 'price']]
df2.head()
```

```python
df2.groupby(['make']).size().plot(kind = "bar")
```

```python
# new dataframe with only the above makes
makes = ['audi', 'bmw', 'mercedes-benz']
df3 = df.loc[df['make'].isin(makes)]
df3['make'].unique()
```

```python
df3.groupby(['make']).sum().plot(kind='pie', y='price')
```
